```python
import json
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
histories = os.listdir('./')
histories = [h for h in histories if h.endswith('.json')]
histories
```

```python
history_dict = {}
for h in histories:
    with open (h) as fp:
        hist = json.load(fp)
        history_dict[h.split('_history')[0]] = hist

# histories[-1].split('_history')
```

```python
def plot_all_histories(how='loss',nrows = 2,figsize=(20,10)):
    ncols = int(np.ceil(len(history_dict)/nrows))
    fig, axs = plt.subplots(nrows=nrows,ncols=ncols,figsize=figsize)
    for nm,data,ax in zip(history_dict.keys(),history_dict.values(),axs.flatten()):
        ax.plot(data[how])
        ax.set_title(nm)

plot_all_histories(how='cosine_similarity')
```

```python
plot_all_histories(how='loss')
```
