```python
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import numpy as np
from deepq import DeepQPlayerV2, visualize, generate_random_board
```

```python
player = DeepQPlayerV2(player_symbol=1)
player.loadPolicy('q_network_q1_same_conv')

tf.keras.utils.plot_model(player.q_network)
```

```python
layer = player.q_network.get_layer('conv2d')
```

```python
def show_weights(layer:tf.keras.layers.Layer,nrows=4,figsize=(20,5)):
    wts = layer.get_weights()[0]
    ncols = int(np.floor(wts.shape[-1]/nrows))
    fig, axs = plt.subplots(nrows=nrows,ncols=ncols,figsize=figsize)

    for wt_idx,ax in zip(range(wts.shape[-1]),axs.flatten()):
        sns.heatmap(wts[:,:,0,wt_idx],ax=ax)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

    
```

```python
show_weights(layer)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import numpy as np
```

```python
def show_weights(layer:tf.keras.layers.Layer,nrows=4,figsize=(20,5),filter_idx=0):
    wts = layer.get_weights()[0]
    ncols = int(np.floor(wts.shape[-1]/nrows))
    fig, axs = plt.subplots(nrows=nrows,ncols=ncols,figsize=figsize)

    for wt_idx,ax in zip(range(wts.shape[-1]),axs.flatten()):
        sns.heatmap(wts[:,:,filter_idx,wt_idx],ax=ax)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
```

```python
mdl = tf.keras.models.load_model('q_network_q2_input_mapper_vsq_small_train')
```

```python
layer = mdl.get_layer('conv2d_2')
```

```python
show_weights(layer,filter_idx=1)
```
