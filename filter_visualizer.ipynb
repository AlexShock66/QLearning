```python
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import numpy as np
from deepq import SinglePredictionQPlayer, visualize, generate_random_board
```

```python
player = SinglePredictionQPlayer(player_symbol=1)
player.loadPolicy('QNetworks/Q_Q_Single_Encoder')

tf.keras.utils.plot_model(player.q_network)
```

```python
layer = player.q_network.get_layer('conv2d')
```

```python
def show_weights(layer:tf.keras.layers.Layer,filter_idx,nrows=4,figsize=(20,5)):
    wts = layer.get_weights()[0]

    ncols = int(np.floor(wts.shape[-1]/nrows))
    fig, axs = plt.subplots(nrows=nrows,ncols=ncols,figsize=figsize)

    for wt_idx,ax in zip(range(wts.shape[-1]),axs.flatten()):
        sns.heatmap(wts[:,:,filter_idx,wt_idx],ax=ax)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

    
```

```python
show_weights(layer,0) #Player 1 cells
```

```python
show_weights(layer,1) #Playe -1 cells
```

```python
show_weights(layer,2) #Empty Cells
```
